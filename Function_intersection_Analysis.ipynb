```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root

# Define the two functions
def y1(x):
    return np.exp(-x)  # Exponential decay function

def y2(x):
    return np.sin(x)   # Sine function

# Define the equation to find where y1 = y2
def equation(x):
    return y1(x) - y2(x)

# Find the root numerically (starting near x=0.5)
result = root(equation, 0.5)
root_x = result.x[0]
print(f"Root found at x â‰ˆ {root_x:.4f}")

# Create x values for plotting
x_vals = np.linspace(-1, 2, 500)

# Plot both functions and the intersection point
plt.figure(figsize=(10, 6), dpi=300)
plt.plot(x_vals, y1(x_vals), label="y1 = exp(-x)", color="blue", linewidth=2)
plt.plot(x_vals, y2(x_vals), label="y2 = sin(x)", color="orange", linewidth=2)
plt.axvline(root_x, color="red", linestyle="--", label=f"Root (x â‰ˆ {root_x:.4f})")
plt.axhline(y1(root_x), color="black",  linestyle="--")
plt.scatter(root_x, y1(root_x), color="red", s=100, zorder=5)

# Add plot formatting
plt.title("Visual Solution of exp(-x) = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
#plt.axhline(0, color="black", linewidth=0.5)
#plt.axvline(0, color="black", linewidth=0.5)
plt.tight_layout()
plt.show()
```
